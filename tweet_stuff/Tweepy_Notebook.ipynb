### Get sentiment for tweets

```python

def get_sentiment(text):
    
    text = text.decode("utf8") # decode if unreadable ascii
    blob = TextBlob(text)
    sent = blob.sentiment.polarity
    return sent

tweets['sentiment'] = tweets['text'].apply(lambda x: get_sentiment(x))
```
